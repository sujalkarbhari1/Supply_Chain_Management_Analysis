```python
import pandas as pd
df = pd.read_csv('SupplyChainManagement_Dataset.csv')
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
```

```python
print(df.columns.tolist())
```

```python
df[['order_date','ship_date','delivery_date']].dtypes
```

```python
date_cols = ['order_date', 'ship_date', 'delivery_date']

for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
df[date_cols].dtypes
```

```python
df['net_sales'] = pd.to_numeric(df['total_sales'], errors='coerce') - \
                  pd.to_numeric(df['discount'], errors='coerce')
```

```python
df['profit_margin%'] = (df['profit'] / df['net_sales']) * 100
```

```python
df['delivery_time'] = (df['delivery_date'] - df['order_date']).dt.days
df['shipping_time'] = (df['ship_date'] - df['order_date']).dt.days
```

```python
df[['total_sales','discount','profit']].dtypes
```

```python
df['return_flag'] = df['return_status'].str.lower().eq('returned')
```

```python
df[['order_date','delivery_date','delivery_time',
    'total_sales','discount','net_sales',
    'profit','profit_margin%']].head()
```

```python
df['order_year'] = df['order_date'].dt.year
df['order_month'] = df['order_date'].dt.month
df['order_month_name'] = df['order_date'].dt.month_name()
df['order_quarter'] = df['order_date'].dt.to_period('Q').astype(str)
```

```python
df.fillna({
    'delivery_time_days': 0,
    'shipping_time_days': 0,
    'profit_margin_pct': 0
}, inplace=True)
```

```python
df.groupby(['order_year','order_month'])['net_sales'].sum().plot(
    title='Monthly Sales Trend'
)
```

```python
df.groupby('category')['profit'].sum().sort_values().plot(kind='barh')
```

```python
df['return_flag'].value_counts().plot(kind='pie', autopct='%1.1f%%')
```
